# Nameplate Annotation Guidelines

## Overview
This guide provides best practices for annotating industrial nameplate images to ensure consistent and high-quality data extraction.

## Key Principles

### 1. Context-Based Key Recognition
When annotating nameplates, **key names may not always be explicitly written**. Use context clues to identify the appropriate field.

**When naming fields, be concise, specific, and consistent.**

**Example:**
- Text shows: "RIMS AT 690 kPa | 100 psi COLD DUAL"
- Context indicates this is tire pressure, measured when cold and in dual configuration
- Use key: `cold_dual_tire_pressure`

### 2. Multiple Values for Single Keys
When a nameplate shows the same measurement in different units, store all values in a list.

**Format:**
```json
"key_name": [
  {"value": 690, "unit": "kPa"},
  {"value": 100, "unit": "psi"}
]
```

### 3. Measurement Data Structure
All measurements must include both value and unit information using this structure:

```json
{"value": number_or_string, "unit": "unit_abbreviation"}
```

## Quick Reference

### When to Use Lists
- Same measurement in multiple units (e.g., weight in kg and lb)

### Special Cases
- **Mixed Units**: Convert to single unit when possible (e.g., 6ft 3in  6'3 ft)

Unit Notation
========================

### Avoid using characters not accessible via classic keyboard

1. MASS
   - lb (pounds)
   - kg (kilograms)
   - ton (tons)
   - g (grams)
   - oz (ounces)

2. LENGTH/DISTANCE
   - m (meters)
   - cm (centimeters)
   - mm (millimeters)
   - km (kilometers)
   - ft (feet) - Note: when given a measurement like 6ft 3in, opt to write 6'3 ft. This maintains the use of only one unit
   - in (inches)

3. PRESSURE
   - kPa (kilopascals)
   - psi (pounds per square inch)
   - bar (bar pressure)
   - MPa (megapascals)

4. ELECTRICAL
   - V (voltage - volts)
   - A or amp (current - amperes)
   - W (power - watts)
   - kW (power - kilowatts)
   - Hz (frequency - hertz)
   - kVA (apparent power - kilovolt-amperes)
   - ohm (resistance - ohms)
   - uF (capacitance - microfarads)

5. TEMPERATURE
   - C (Celsius degrees)
   - F (Fahrenheit degrees)
   - K (Kelvin)

6. VOLUME
   - L (liters)
   - gal (gallons)
   - m3 (cubic meters)

7. FLOW RATE
   - L/min (liters per minute)
   - gal/min (gallons per minute)
   - m3/h (cubic meters per hour)

8. SPEED/ROTATION
   - rpm (revolutions per minute)
   - Hz (frequency - hertz)
